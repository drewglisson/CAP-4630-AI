### Setting up data and equation

```
import numpy as np

%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rc('axes', labelsize=14)
mpl.rc('xtick', labelsize=12)
mpl.rc('ytick', labelsize=12)
```

```
# generate m random pairs (x, y)

m = 100
x1 = 2 * np.random.rand(m, 1)
x2 = 4 * np.random.rand(m, 1)
y = x1 ** x2 + np.random.rand(m, 1)
```

```

plt.plot(x1, y, "b.")
plt.plot(x2, y, "r.")

plt.xlabel("$x_1, x_2$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.axis([0, 2, 0, 15])
plt.show()
```

### Mini Batch 

```
# number of epochs
epochs = 20
# learning rate
lr = 0.01
# fix initial random weight for gradient descent (batch gradient descent)
np.random.seed(42)
initial_weight = np.random.randn(3, 1) 
print(initial_weight)
```

```
# sanity check (I will use this to explain how the vectorized code works)
print(initial_weight.shape)
print(y.shape)
print((X_b.dot(initial_weight)).shape)
print((X_b.T.dot(X_b.dot(initial_weight))).shape)
```

```
# plot training data
# plt.plot(X, y, "b.")

weight = initial_weight
weight_path_mgd = []
batch_size = 4

weight_path_mgd.append(weight)
for epoch in range(epochs):
    shuffled_indices = np.random.permutation(m )
    X_b_shuffled = X_b[shuffled_indices]
    y_shuffled = y[shuffled_indices]
    for i in range(0, m, batch_size):
        xi = X_b_shuffled[i:i+batch_size]
        yi = y_shuffled[i:i+batch_size]

        gradient = 1 / batch_size * xi.T.dot(xi.dot(weight) - yi)
        weight = weight - lr * gradient
        weight_path_mgd.append(weight)
        print(weight.shape)                    

        y_predict = X_new_b.dot(weight)
        plt.plot(X_new, y_predict, "b-")  
      
plt.xlabel("$x_1$", fontsize=18)                     
plt.ylabel("$y$", rotation=0, fontsize=18)           
plt.axis([0, 2, 0, 15])                                              
plt.show() 
```

## Problem 3

What does an average MNIST digit look like? For each i=0,1,...,9, compute the average of digit i and display it. More precisely, you have add all the images of the digit i together and divide it by the number of times the digit i occurs in the data set.

Source: https://colab.research.google.com/github/schneider128k/machine_learning_course/blob/master/mnist_digits_data_set.ipynb#scrollTo=UB7_xf0hzsPx

### Load the data set

```
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

%tensorflow_version 2.x
import tensorflow as tf

(train_images, train_labels), (test_images, test_labels) = tf.keras.datasets.mnist.load_data()
```

### Plot the average of each digit

```
average = np.zeros((10,28,28), dtype=np.int)
counter = np.zeros(10, dtype=np.int)

for i in range(6000):
  counter[train_labels[i]] += 1
  average[train_labels[i]] = average[train_labels[i]] + train_images[i]

plt.figure(figsize=(16, 16))
for n in range(10):
  average[n] = average[n] / counter[n]
  ax = plt.subplot(4, 4, n+1)
  plt.imshow(average[n])
  plt.title(n)
  plt.axis('off')
```
