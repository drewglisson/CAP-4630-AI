## Multiplying Matrices

In order to multiply matrices, the number of columns in the first matrix must be equal to the number of rows in the second. The product matrix has the number of rows of the first and the number of columns of the second. In the figure below, there are two matrices, 2x3 and 3x2, we can multiply these two because the rows in the first match the columns in the second. <p>
We multiply them by using the dot product. According to the figure below, we take the dot product of the first row of the first matrix and the first column of the second matrix, and place that in the top left spot in the product. Then take the product of the same row with the next column in the second matrix and continue until the last column in the second. Then you move onto the second row in the first matrix and do the same until the last row in the first matrix.


$$\begin{bmatrix}
    a  &  b  &  c     \\
    d  &  e  &  f
\end{bmatrix}
\times
\begin{bmatrix}
    g  &  h      \\
    i  &  j      \\
    k  &  l
\end{bmatrix}
=
\begin{bmatrix}
    (a \times g) + (b \times i) + (c \times k) &       (a \times h) + (b \times j) + (c \times l) \\      (d \times g) + (e \times i) + (f \times k) &  
    (d \times h) + (e \times j) + (f \times l) \\
\end{bmatrix}$$

## multiply_matrices
This function takes an arbitrary number of matrices and multiplies these together to find the matrix product. The program goes through the list of matrices and multiples them two at a time. First it multiplies array 0 and array 1, then multiply that product with array 2, until the list is complete. I utilize python's @ operator to multiply the matrices. <p>
The program prints out the final product for the list of matrices. it also raises an exception when there is an empty array or if two arrays cannot be multiped. 


```
import numpy as np

class Error(Exception):
  """Base class for other exceptions"""
  pass

class NotEqual(Error):
  """Raised when the matrices the number of columns in x does not equal the number of rows in y"""
  pass

class Empty(Error):
  """Raised when there is an empty matrix"""
  pass

class OnlyOne(Error):
  """ Raised when there only one array """
  pass 

def multiply_matrices(*arrays):

  j = 1
  s = 0
  x = arrays[0]

  for i in arrays:
    try: 
      if len(arrays) == 1:
        raise OnlyOne

    except OnlyOne:
      print("Exception: there is only one array, please add more")
      return
    y = arrays[j]
    
    # exceptions 
    try: 
      xcolumns = np.size(x,1)
      # xrows = np.size(x,0)
      yrows = np.size(y,0)
      # ycolumns = np.size(y,1)
      
      if not x.any():
        empty = x
        raise Empty

      if not y.any():
        empty = j
        raise Empty
        
      if xcolumns != yrows:
        raise NotEqual

      if len(arrays) == 1:
        raise OnlyOne

    except NotEqual:
      print("Exception on array : The arrays cannot be multipied,")
      print('the rows of matrix', s, 'do not match the columns of matrix', j)
      return
    
    except Empty:
      print("Exception: Matrix", empty,"is empty, cannot mulitpy")
      return
    
    x = x @ y
    if j < (len(arrays) - 1):
      j+=1
    else:
      break

    s+=1

  print(x)
  return 
```

## Test Cases
There are four test cases implemented. Multiplying two matrices with same rank, multiplying two matrices with different rank, multiplying three matrices with same rank, and multiplying a full matrix with an empty matrix. <p>
The first and third cases will output the product of the arrays. The second case will raise an exception that the matrices are not compatible to multiply. Finally, the last case will raise the expectation that there is an empty matrix. 


```
import numpy as np

array1 = np.full((4,7), 20)
array2 = np.full((7,5), 5)

print("-------Multiping two matrices with same rank")
multiply_matrices(array1, array2)
print()

array3 = np.full((4,7), 20)
array4 = np.full((3,5), 5)

print("-------Multiping two matrices with different rank")
multiply_matrices(array3, array4)
print()

array5 = np.full((4,7), 20)
array6 = np.full((7,5), 5)
array7 = np.full((5,9), 8)

print("-------Multiping three matrices with same rank")
multiply_matrices(array5, array6, array7)
print()

array8 = np.full((4,7), 20)
array9 = []

print("-------Multiping a full matrix and an empty matrix")
multiply_matrices(array8, array9)
print()

print("-------Multiping a full matrix and an empty matrix")
multiply_matrices(array8)
```
